Este notebook simula o problema de atualização de arquivos em containers Docker e demonstra a solução.

---

## 1. Criando estrutura do projeto

```python
import os

os.makedirs('app', exist_ok=True)

with open('app/index.html', 'w') as f:
    f.write('<h1>Versão 1</h1>')

with open('app/run.py', 'w') as f:
    f.write('''from http.server import SimpleHTTPRequestHandler, HTTPServer
PORT = 8000
with HTTPServer(("0.0.0.0", PORT), SimpleHTTPRequestHandler) as httpd:
    print("Servidor rodando")
    httpd.serve_forever()''')

print('Estrutura criada')
```

## 2. Criando Dockerfile

```python
dockerfile = '''FROM python:3.9-slim
WORKDIR /app
COPY app/ /app/
CMD ["python", "run.py"]'''

with open('Dockerfile', 'w') as f:
    f.write(dockerfile)

print('Dockerfile criado')
```

## 3. Criando docker-compose.yml

```python
compose = '''version: '3'
services:
  web:
    build: .
    ports:
      - "8080:8000"
    volumes:
      - logs:/log

volumes:
  logs:
'''

with open('docker-compose.yml', 'w') as f:
    f.write(compose)

print('docker-compose criado')
```

## 4. Simulação do problema

Alterando arquivo no host...

```python
with open('app/index.html', 'w') as f:
    f.write('<h1>Versão 2</h1>')

print('Arquivo alterado no host!')
```

## 5. Simulação do comportamento do container

Como o Docker usa COPY, o container ainda mantém versão antiga.

```python
print('Conteúdo esperado dentro do container:')
print('<h1>Versão 1</h1>')
```

## 6. Solução com Bind Mount

```python
compose_fix = '''version: '3'
services:
  web:
    build: .
    ports:
      - "8080:8000"
    volumes:
      - ./app:/app
      - logs:/log

volumes:
  logs:
'''

with open('docker-compose.yml', 'w') as f:
    f.write(compose_fix)

print('Bind mount aplicado!')
```

## 7. Resultado esperado

Agora alterações no host refletem imediatamente no container.
