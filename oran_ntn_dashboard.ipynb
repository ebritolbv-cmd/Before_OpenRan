```python

import random
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python

def generate_metrics():
    return {
        "timestamp": datetime.utcnow(),
        "throughput": random.uniform(80, 150),
        "latency": random.uniform(50, 120),
        "jitter": random.uniform(1, 10),
        "packetLoss": random.uniform(0, 2),
        "satelliteQuality": random.uniform(85, 100),
        "sinr": random.uniform(10, 25),
        "ptpOffset": random.uniform(50, 500),
        "connections": random.randint(20, 80)
    }
```

```python

data = []
for _ in range(50):
    data.append(generate_metrics())

import pandas as pd
df = pd.DataFrame(data)
df.head()
```

```python

import matplotlib.pyplot as plt

plt.figure()
plt.plot(df['throughput'])
plt.title('Throughput (Mbps)')
plt.xlabel('Tempo')
plt.ylabel('Mbps')
plt.show()
```

```python

plt.figure()
plt.plot(df['latency'])
plt.title('Latência (ms)')
plt.xlabel('Tempo')
plt.ylabel('ms')
plt.show()
```

```python

plt.figure()
plt.plot(df['satelliteQuality'])
plt.title('Qualidade do Satélite (%)')
plt.xlabel('Tempo')
plt.ylabel('%')
plt.show()
```

```python

summary = {
    "Throughput Médio": df['throughput'].mean(),
    "Latência Média": df['latency'].mean(),
    "Jitter Médio": df['jitter'].mean(),
    "Packet Loss Médio": df['packetLoss'].mean(),
}
summary
```
